Widgets are basic User Interface (UI) building blocks that maintain a state and can trigger changes using callbacks.

The easiest way to bind some interactivity to a widget is to use the `@interact` decorator on a function:

```python
from ipywidgets import interact
```

```python
import spacy
from spacy import displacy
```

```python
nlp = spacy.load("en_core_web_sm")
```

### Intermezzo: decorators

```python
def analyze(text, a):
    displacy.render(nlp(text), jupyter=True)
```

```python
misterious_object = interact(text="", a=True)
```

```python
misterious_object(analyze)
```

```python
def my_wrapper(function):
    def wrapper():
        print("I am going to call the function:")
        function()
        print("I called the function")

    return wrapper
```

```python
def func():
    print("42")
```

```python
func()
```

```python
new_func = my_wrapper(func)
```

```python
new_func()
```

```python
@my_wrapper
def func():
    print("42")
```

```python
func()
```

```python
#@interact(text="", a=True)
def analyze(text):
    displacy.render(nlp(text), jupyter=True)
```

```python
import pandas as pd
```

```python
fertility = pd.read_csv("data/gapminder/fertility.csv", index_col="Country")
fertility.head()
```

```python
population = pd.read_csv("data/gapminder/population.csv", index_col="Country")
population.head()
```

```python
life_exp = pd.read_csv("data/gapminder/life_expectancy.csv", index_col="Country")
life_exp.head()
```

```python
regions = pd.read_csv("data/gapminder/regions.csv", index_col="Country")
regions.head()
```

```python
def by_year_alt(year):
    df = pd.concat([
        population[str(year)],
        fertility[str(year)],
        life_exp[str(year)],
        regions['Group']
    ], axis=1, sort=True)
    df.columns = ["P", "F", "LE", "G"]
    
    return df
```

```python
def by_year(year):
    return pd.DataFrame({
        'Population': population[str(year)],
        'Fertility': fertility[str(year)],
        'Life expectancy': life_exp[str(year)],
        'Group': regions['Group'],
    })
```

```python
df = by_year(1991)
df.head()
```

```python
fig = go.FigureWidget()
fig
```

```python
df.head()
```

```python
for a, b in df.groupby("Group"):
    print(a, type(b))
```

```python
for group_name, sub_df in df.groupby("Group"):
    sc = fig.add_scatter(
        x=sub_df['Fertility'],
        y=sub_df['Life expectancy'],
        mode='markers',
        marker={
            'size': np.sqrt(sub_df['Population'].fillna(0))/ 400
        },
        name=group_name,
    )
```
